```python
#importing libraries
import numpy as np
import pandas as pd
data = pd.read_csv("instagram_global_top_1000.csv")
```

```python
data.isnull().sum()
```

```python
data = data.fillna({'Audience Country':'United States'})
```

```python
data = data.fillna({'Category':'Music|Modeling'})
```

```python
data=data.fillna(method = 'pad',axis = 1)
```

```python
data.drop(["Country", "Scraped"], axis = 1, inplace = True)
```

```python
data = data.astype({"Followers":"int","Authentic engagement":"int","Engagement avg":"int"})
```

```python
data.pivot_table(index="Category", values="Followers", aggfunc=['mean', 'sum', 'min', 'max'], sort=False)
```

```python
data.pivot_table(index="Category", values="Engagement avg", aggfunc=['mean', 'sum', 'min', 'max'], sort=False)
```

```python
# returning the relative frequency of audience country elements
data['Audience Country'].value_counts(normalize=True).round(2)
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns 
plt.figure(figsize=(35,10))
sns.countplot(x="Audience Country", data = data)
plt.title("Relative Frequency of Audience Country",fontsize =20)
plt.xlabel("Audience Country",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```
